```python
#Eliminar primera fila o filas innecesarias
df= pd.read_excel('Gastosycostos20-23.xlsx', skiprows=5)
df.head(5)
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
df2=df.copy()
```

Reemplacé los valores nulos con el método backward fill (bfill), o sea rellené con el dato siguiente válido. Lo hice porque el folio es un identificador que no puede quedar vacío así se mantiene una secuencia.

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
df2["FOLIO"] =df2["FOLIO"].fillna(method="bfill") 
```

Sustituí los nulos con el numero 0 porque existen registros con 0 entonces los no registrados se llenaran con el mismo numero para no afectar el resultado y para que no queden con espacios vacíos.

```python
df2["GASTO"] =df2["GASTO"].fillna(0)
```

Reemplacé los valores nulos con 1 por que se puede asumir el valor base de 1 ya que ya se implemento en el excel de esa manera.

```python
df2["TC"]= df2["TC"].fillna(1) 
df2
```

Los valores nulos los sustituí con la media del importe para mantener un valor representativo.

```python
df2["IMPORTE"]=df2["IMPORTE"].fillna(round(df["IMPORTE"].mean(),1))
```

También usé la media para reemplazar los nulos para consistencia y da un valor intermedio representativo de la columna.

```python
df2["IVA"]=df2["IVA"].fillna(round(df["IVA"].mean(),1))
```

Los valores nulos los sustituí con el texto “no poliza”, porque así queda claro que en ese registro no se le aplicó ninguna póliza

```python
df2["POLIZA"] =df2["POLIZA"].fillna("no poliza")
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
df3= df2.dropna()
df3
```

Elimine los valores nulos de la variable TIPO ya que solo tenia 1 valor nulo

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```
