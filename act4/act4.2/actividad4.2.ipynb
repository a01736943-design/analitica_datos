```python
#Importamos librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('listings.csv')
df.info()
```

```python

filtro_a= df["host_name"].value_counts().head(5).index.tolist()
filtro_a = df[df["host_name"].isin(filtro_a)]
filtro_a.head()
filtro_a.to_csv("filtro_a.csv")

```

```python
filtro_b = df[df["host_since"] > "2020-12-31"] 
filtro_b.to_csv("filtro_b.csv")
filtro_b
```

```python
filtro_c = df[df["host_response_time"] == "within a day"]
filtro_c.to_csv("filtro_c.csv")
```

```python
filtro_d= df.iloc[200::200,:] #Filas con incrementos de 200 a partir de la fila 200
filtro_d
filtro_d.to_csv("filtro_d.csv")
```

```python
filtro_e = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtro_e.to_csv("filtro_e.csv")
```

```python
filtro_f = df[df["beds"] <= 4]
filtro_f.to_csv("filtro_f.csv")
```

```python
filtro_g = df[df["host_is_superhost"] == "t"]
filtro_g.to_csv("filtro_g.csv")
```

```python
#Columnas incrementales en un fator
filtro_h= df.iloc[:, 1::2] #Columnas con incrementos de 2 a partir de la columna
filtro_h.to_csv("filtro_h.csv")
```
