Generar una librería que incluya las siguientes funciones personalizadas:
a) Carga de archivos (csv, xlsx, html y json) a dataframes .
b) Sustitución de valores nulos con el método de “ffill” (Para todas las variables).
c) Sustitución de valores nulos con el método de “bfill” (Para todas las variables).
d) Sustitución de valores nulos con el método por un “string concreto” (Solo para variables String).
e) Sustitución de valores nulos con el método de “promedio” (Solo para variables Numéricas).
f) Sustitución de valores nulos con el método de “mediana” (Solo para variables Numéricas).
g) Sustitución de valores nulos con el método de “constante” (Solo para variables Numéricas).
i) Identificación de valores nulos por columna y por dataframe” (Para todas las variables).

```python
# Importamos librerías estándar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Importamos nuestra librería
import Ania

# CARGA DE ARCHIVO DE PRUEBA
df = Ania.cargar_dataset("listings.csv")

print("Primeras filas del archivo")
print(df.head())

# IDENTIFICACIÓN DE NULOS  i)
print("Nulos por columna y total DENTIFICACIÓN DE NULOS  i)")
print(Ania.identificar_nulos(df))

# LIMPIEZA DE NULOS - FFILL b
df_ffill = Ania.nulos_ffill(df)
print("Nulos después de aplicar ffill LIMPIEZA DE NULOS - FFILL b")
print(df_ffill.isnull().sum())

# LIMPIEZA DE NULOS - BFILL c
df_bfill = Ania.nulos_bfill(df)
print("Nulos después de aplicar bfill LIMPIEZA DE NULOS - BFILL c")
print(df_bfill.isnull().sum())

# LIMPIEZA DE NULOS - STRING CONCRETO d
df_string = Ania.nulos_string(df, valor="SIN DATO")
print("Nulos después de aplicar string concreto  LIMPIEZA DE NULOS - STRING CONCRETO d")
print(df_string.isnull().sum())

# LIMPIEZA DE NULOS - PROMEDIO e
df_prom = Ania.nulos_prom_df(df)
print("Nulos después de aplicar promedio LIMPIEZA DE NULOS - PROMEDIO e")
print(df_prom.isnull().sum())

# LIMPIEZA DE NULOS - MEDIANA f
df_median = Ania.nulos_median_df(df)
print("Nulos después de aplicar mediana LIMPIEZA DE NULOS - MEDIANA f")
print(df_median.isnull().sum())

# LIMPIEZA DE NULOS - CONSTANTE g
df_constante = Ania.nulos_constante_df(df, constante=999)
print("Nulos después de aplicar constante LIMPIEZA DE NULOS - CONSTANTE g")
print(df_constante.isnull().sum())

string = Ania.string(df)
print("string")
print(string.isnull().sum())
```
